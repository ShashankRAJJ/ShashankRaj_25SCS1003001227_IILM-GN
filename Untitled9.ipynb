```python
import os, time, re, traceback
import cv2
import numpy as np
import pytesseract

# ensure folder exists
os.makedirs("Detected_Plates", exist_ok=True)

# tesseract path (Colab default) - adjust if needed
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

# load cascade and check
plate_cascade = cv2.CascadeClassifier("haarcascade_russian_plate_number.xml")
if plate_cascade.empty():
    raise RuntimeError("Failed to load haarcascade_russian_plate_number.xml. Download or fix path.")

states = {
    "AN": "Andaman and Nicobar", "AP": "Andhra Pradesh", "AR": "Arunachal Pradesh",
    "AS": "Assam", "BR": "Bihar", "CH": "Chandigarh", "CG": "Chhattisgarh",
    "DN": "Dadra and Nagar Haveli", "DD": "Daman and Diu", "DL": "Delhi",
    "GA": "Goa", "GJ": "Gujarat", "HR": "Haryana", "HP": "Himachal Pradesh",
    "JK": "Jammu and Kashmir", "JH": "Jharkhand", "KA": "Karnataka",
    "KL": "Kerala", "LA": "Ladakh", "LD": "Lakshadweep", "MP": "Madhya Pradesh",
    "MH": "Maharashtra", "MN": "Manipur", "ML": "Meghalaya", "MZ": "Mizoram",
    "NL": "Nagaland", "OD": "Odisha", "PY": "Puducherry", "PB": "Punjab",
    "RJ": "Rajasthan", "SK": "Sikkim", "TN": "Tamil Nadu", "TS": "Telangana",
    "TR": "Tripura", "UP": "Uttar Pradesh", "UK": "Uttarakhand", "WB": "West Bengal"
}

def _resize_max_side(img, max_side=1280):
    h,w = img.shape[:2]
    if max(h,w) <= max_side:
        return img
    scale = max_side / max(h,w)
    return cv2.resize(img, (int(w*scale), int(h*scale)), interpolation=cv2.INTER_AREA)
```

```python
def detect_plate(image: np.ndarray):
    """
    Input: image as RGB numpy array (gradio's type='numpy')
    Output: (text_str, cropped_plate_rgb_or_None)
    """
    try:
        if image is None:
            return "No image provided", None

        # 1) ensure size is reasonable (avoid OOM)
        img_rgb = _resize_max_side(image.copy(), max_side=1280)

        # 2) grayscale from RGB (correct conversion)
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

        # 3) detect plates - Adjusted parameters here
        plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=5, minSize=(30,30))

        outputs = []
        last_crop_rgb = None

        for (x, y, w, h) in plates:
            # add small padding
            pad = int(0.03 * max(w, h))
            x1, y1 = max(x - pad, 0), max(y - pad, 0)
            x2, y2 = min(x + w + pad, img_rgb.shape[1]-1), min(y + h + pad, img_rgb.shape[0]-1)

            crop_rgb = img_rgb[y1:y2, x1:x2].copy()
            last_crop_rgb = crop_rgb

            # preprocessing for OCR
            gray_crop = cv2.cvtColor(crop_rgb, cv2.COLOR_RGB2GRAY)
            gray_crop = cv2.bilateralFilter(gray_crop, 9, 75, 75)
            # try Otsu
            _, thresh = cv2.threshold(gray_crop, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

            # OCR: whitelist alphanumerics, psm 7 (single line)
            config = "--oem 3 --psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
            text = pytesseract.image_to_string(thresh, config=config)
            clean = re.sub(r'[^A-Za-z0-9]', '', text).upper()

            # fallback: try a different psm if too short
            if len(clean) < 3:
                config2 = "--oem 3 --psm 11 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
                text2 = pytesseract.image_to_string(thresh, config=config2)
                clean2 = re.sub(r'[^A-Za-z0-9]', '', text2).upper()
                if len(clean2) > len(clean):
                    clean = clean2

            if len(clean) < 3:
                continue

            state = states.get(clean[:2], "Unknown")
            timestamp = time.strftime("%Y%m%d_%H%M%S")
            filename = f"{clean}_{timestamp}.jpg"

            # cv2.imwrite expects BGR
            crop_bgr = cv2.cvtColor(crop_rgb, cv2.COLOR_RGB2BGR)
            cv2.imwrite(os.path.join("Detected_Plates", filename), crop_bgr)

            outputs.append(f"Plate: {clean} | State: {state} | Saved: {filename}")

        if not outputs:
            return "No plate detected", None

        return "\n".join(outputs), last_crop_rgb

    except Exception as e:
        # optional: write full traceback to file for debugging
        with open("/content/detect_plate_error_log.txt", "a") as f:
            f.write(time.strftime("%Y-%m-%d %H:%M:%S") + "\n")
            traceback.print_exc(file=f)
            f.write("\n")
        return "Internal error occurred — check /content/detect_plate_error_log.txt", None
```

```python

import gradio as gr

interface = gr.Interface(
    fn=detect_plate,
    inputs=gr.Image(type="numpy", label="Upload Car Image"),
    outputs=[
        gr.Textbox(label="Detected Information"),
        gr.Image(label="Cropped Plate")
    ],
    title="Number Plate Detection (Google Colab)",
    description="Upload a car image to detect and extract number plate."
)

interface.launch(debug=True)
```
